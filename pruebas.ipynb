```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import math
```

```python
golden = (1 + 5 ** 0.5) / 2
mpl.rcParams.update({'font.size': 12, 'text.usetex': True})
```

```python
def procesar(dimensionA, dimensionB, dimensionC,totalTime=10, timeSteps=100,Lambda=.1,theta=math.pi/2,gamma=0.1):
    datos = !./mein -o get_purity --theta $theta --gamma $gamma --lambda $Lambda --time_steps $timeSteps --dimension-c $dimensionC --dimension-b $dimensionB  --dimension-a $dimensionA --seed 0 --no_general_report  --total_time $totalTime
    datos_nuevos = []
    for line in datos:
        datos_nuevos.append( map(float, line.split()) )
    datos_nuevos = np.array(datos_nuevos)
    return datos_nuevos
```

```python
def four_plots(a):
    fig, axes = plt.subplots(1, 4, figsize=(12, 4))
    axes[0].plot(a[:,0], a[:,3])
    axes[0].set_title(r"$P(\rho_A)$")
    axes[1].plot(a[:,0], a[:,2])
    axes[1].set_title(r"$P(\rho_B)$")
    axes[2].plot(a[:,0], a[:,1])
    axes[2].set_title(r"$P(\rho_C)$")
    axes[3].plot(a[:,0], a[:,1], label=r"$P(\rho_C)$")
    axes[3].plot(a[:,0], a[:,2], label=r"$P(\rho_B)$")
    axes[3].plot(a[:,0], a[:,3], label=r"$P(\rho_A)$")
    axes[3].legend(prop={'size':10})
```

# Primera prueba $P_A =? P_B P_C$? Quiza es una aplicacion de la formula de quantum memories nuestra

```python
dimensionC=2;
dimensionB=4;
dimensionA=20;
totalTime=10;
timeSteps=200;
Lambda=0.1;
theta=1.57;
gamma=1.;
a=procesar(dimensionA, dimensionB, dimensionC,totalTime,timeSteps,Lambda,theta,gamma);
```

```python
plt.clf();
s=2;
fig=plt.figure(figsize=((golden*s,s)), dpi=300)
plt.plot(a[:,0], a[:,1], label=r"$P(\rho_C)$")
plt.plot(a[:,0], a[:,2], label=r"$P(\rho_B)$")
plt.plot(a[:,0], a[:,3], label=r"$P(\rho_A)$")
plt.legend(prop={'size':6})
plt.xlabel("$t$");
plt.ylabel("$P$");
plt.ylim(0, 1);
```

# Para $\lambda$ chico, c debe ser puro y $\theta=0$ A y B no se deben mezclar

```python
# Graficar con lambda chico, c debe ser puro, para theta=0, a y B deben ser puros
# Gamma deberia ser irrelevante. 
# H_env = cos theta (H_A + H_B) + \gamma \sin \theta V_{AB}
# H_cen = H_C
# H_{int} = V_{BC}
# H  = H_{env} + H_{cen} + \lambda H_int
```

```python
dimensionC, dimensionB, dimensionA=2, 4, 40;
Lambda,theta, gamma=0.001,0,1;
a=procesar(dimensionA, dimensionB, dimensionC,totalTime,timeSteps,Lambda,theta,gamma);
four_plots(a)
```

```python
dimensionC, dimensionB, dimensionA=2, 4, 40;
Lambda,theta, gamma=0.01,0,1;
a=procesar(dimensionA, dimensionB, dimensionC,totalTime,timeSteps,Lambda,theta,gamma);
four_plots(a)
```

# Para $\lambda$ chico, c debe ser puro y $\theta=\pi/2$ A y B si se deben mezclar

```python
dimensionC, dimensionB, dimensionA=2, 4, 40;
Lambda,theta, gamma=0.001,pi/2,1;
totalTime=10;
a=procesar(dimensionA, dimensionB, dimensionC,totalTime,timeSteps,Lambda,theta,gamma);
four_plots(a)
```
