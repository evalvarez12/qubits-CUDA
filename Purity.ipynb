## Variando el tamaño del ambiente

```python
#CAOS EN CADENA
Jc,Js,DJs,Dbs,bx,bz=[0.05,0.8,0.1,0,0.8,0.8] 
```

```python
qubits=[2,4,6,8,10,12]
```

```python
for q in qubits :
    dat_q=str(q)+".dat"
    !rm purity$dat_q
    # Todos se realiza muchas veces para obtener un promedio
    promedio=5
    for i in range(promedio) :
        !ssh t800  carlospg/abc/./abc -o purity -q $q --model model1 --Jc $Jc --Js $Js --DJs $DJs --bx $bx --bz $bz --Dbs $Dbs --t 400 --PARAMseed 23343 --Eseed 2456>> purity$dat_q

 
```

```python
#yscale('log')
for q in qubits :
    dat_q=str(q)+".dat"
    purity_all=loadtxt("purity"+dat_q)
    purity=zeros(len(purity_all[0]))
    for i in purity_all :
        purity += i
    purity=purity/len(purity_all)    
    plot(purity,label=str(q))
legend()
ylabel('$\gamma$')
xlabel('$t$')
axis([0,100,.48,1])
```

Viendo para un cuadro de tiempo

```python
yscale('log')
t=5
for q in qubits :
    dat_q=str(q)+".dat"
    purity_all=loadtxt("purity"+dat_q)
    purity=zeros(len(purity_all[0]))
    for i in purity_all :
        purity += i
    purity=purity/len(purity_all)    
    plot(q,purity[t]-.5,'bo')
ylabel('$\gamma$')
xlabel('$n$')
#axis([4,18,.48,1])
```
