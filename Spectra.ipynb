## Se calcula la $P(s)$

```python
dist=zeros(len(spectra))
for i in range(len(spectra)-1) :
    dist[i]=abs(spectra[i]-spectra[i+1])
dist[len(spectra)-1]=abs(spectra[0]-spectra[-1]+2*pi)  
dist=dist/average(dist)
def goe(x) :
  return pi*x/2*exp(-pi*x**2/4)

def poiss(x) :
  return exp(-x)
i=linspace(0,2*pi,100)
plot(i,goe(i))
plot(i,poiss(i))
hist(dist,50,range=(0,2*pi),normed=True,histtype='step')
ylabel('$p(s)$')
xlabel('$s$')
```

## Spectro usando simetrias de reflexion

```python
#Los parametros del sistema:
Jc,Js,DJs,Dbs,bx,bz=[0,1.,0.,0.,1.,1.]
```

```python
!ssh t800  carlospg/abc/./abc -o get_spectra -q 10 --model chain_open --Jc $Jc --Js $Js --DJs $DJs --Dbs $Dbs --bx $bx --bz $bz --symR 1 --PARAMseed 23343  > spectra1.dat
!ssh t800  carlospg/abc/./abc -o get_spectra -q 10 --model chain_open --Jc $Jc --Js $Js --DJs $DJs --Dbs $Dbs --bx $bx --bz $bz --symR -1 --PARAMseed 23343  > spectra2.dat
spectra1=loadtxt("spectra1.dat")
spectra2=loadtxt("spectra2.dat")

spectraxy=sort(spectra1)
y,x=histogram(spectraxy,60,normed=True)
x=x[:-1]
axis([0,2*pi,0,1])
ylabel('$P(\phi$)')
xlabel('$\phi$')
plot(x,y)


```

```python
def goe(x) :
  return pi*x/2*exp(-pi*x**2/4)

def poiss(x) :
  return exp(-x)
ii=linspace(0,2*pi,100)
plot(ii,goe(ii))
plot(ii,poiss(ii))


tdist=[]
for spectra in [spectra1,spectra2] :
    spectra=sort(spectra)
    dist=zeros(len(spectra))
    for i in range(len(spectra)-1) :
        dist[i]=abs(spectra[i]-spectra[i+1])
    dist[len(spectra)-1]=abs(spectra[0]-spectra[-1]+2*pi)  
    tdist+=dist.tolist()

tdist=array(tdist)
tdist=tdist/average(tdist)
hist(tdist,50,range=(0,2*pi),normed=True,histtype='step')
ylabel('$p(s)$')
xlabel('$s$')
```
